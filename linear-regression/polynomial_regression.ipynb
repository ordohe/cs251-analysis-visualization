```python
# Initialize an empty list to store MSE values
mse_values = []

# Perform polynomial regression for degrees 1 to 7
for degree in range(1, 8):
    # Perform polynomial regression with the current degree
    lin_reg.poly_regression('X', 'Y', degree)
    
    # Calculate the MSE
    mse = lin_reg.mse
    
    # Append the MSE to the list
    mse_values.append(mse)

# Plot the change in MSE
plt.figure(figsize=(10, 6))
plt.plot(range(1, 8), mse_values, marker='o')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Change in MSE with Polynomial Degree')
plt.grid(True)
plt.xticks(range(1, 8))
plt.show()
```

#### Check overfitting with $p = 7$ polynomial regression model

In the cell below:

```python
# Fit polynomial regression with p=7
p = 1
fit_lr = linear_regression.LinearRegression(fit)
fit_lr.poly_regression('X', 'Y', int(p))


# New LinearRegression with validation data
val_lr = linear_regression.LinearRegression(val)
val_lr.poly_regression('X', 'Y', int(p))
val_lr.initialize('X', 'Y', fit_lr.get_fitted_slope(), fit_lr.get_fitted_intercept(), p)

# Scatterplots
fit_lr.scatter('X', 'Y', 'Fit Set')
val_lr.scatter('X', 'Y', 'Validation Set')

# Compute MSE
fit_mse = fit_lr.compute_mse()
val_mse = val_lr.compute_mse()
print('Fit MSE:', fit_mse)
print('Validation MSE:', val_mse)
```
